## PROCESS STARTS

## Import Required Libraries

```python
import pandas as pd
```

Pandas is used for data manipulation and preprocessing.


## Load Clean Dataset

```python
df = pd.read_csv("train_ctrUa4K.csv")
df_clean = df.copy()
```

The dataset is loaded using pandas. A copy of the dataset is created to perform encoding operations without modifying the original data.

## Identify Categorical Features

```python
df_clean.select_dtypes(include=['object']).columns
```

This command selects columns that contain categorical values (text data). These columns cannot be directly used by machine learning models and must be encoded into numerical format. As you can see there are 8 categorical columns.

## Label Encoding for Binary Categories

```python
df_encoded = df_clean.copy()

df_encoded['Gender'] = df_encoded['Gender'].map({'Male':1, 'Female':0})
df_encoded['Married'] = df_encoded['Married'].map({'Yes':1, 'No':0})
df_encoded['Education'] = df_encoded['Education'].map({'Graduate':1, 'Not Graduate':0})
df_encoded['Self_Employed'] = df_encoded['Self_Employed'].map({'Yes':1, 'No':0})
df_encoded['Loan_Status'] = df_encoded['Loan_Status'].map({'Y':1, 'N':0})
```

Label encoding converts binary categorical values into numerical values.

Examples:
Male → 1, Female → 0  
Yes → 1, No → 0  

This is appropriate for binary categories because there are only two possible values.

## One-Hot Encoding for Property_Area

```python
df_encoded = pd.get_dummies(df_encoded, columns=['Property_Area'], drop_first=True)
```

Property_Area contains multiple categories (Rural, Semiurban, Urban) with no natural order.

One-hot encoding creates separate binary columns for each category.

The parameter drop_first=True prevents the dummy variable trap by removing one redundant column.

## Convert Dependents Column to Numeric

```python
df_encoded['Dependents'] = df_encoded['Dependents'].fillna(df_encoded['Dependents'].mode()[0])
df_encoded['Dependents'] = df_encoded['Dependents'].replace('3+',3)
df_encoded['Dependents'] = df_encoded['Dependents'].astype(int)
```

The Dependents column contains values such as 0, 1, 2, and '3+'.

Steps performed:
1. Missing values are filled using the mode.
2. '3+' is replaced with 3.
3. The column is converted to integer type.

This allows the column to be used as a numeric feature.

## Remove Non-Predictive Feature

```python
df_encoded = df_encoded.drop('Loan_ID',axis=1)
```

Loan_ID is only an identifier and does not contribute to predicting loan approval.

Keeping such columns may introduce noise into the model, so it is removed from the dataset.

## Verify Dataset After Encoding

```python
df_encoded.select_dtypes(include=['object']).columns
```

This step ensures that no categorical columns remain. The output should return an empty list, indicating that all features are numeric.

## Dataset Shape After Encoding

```python
df_encoded.shape
```

The number of columns may increase after one-hot encoding because new binary columns are created. And due to dataset calculations this is perfect.

# Final Observations

• All categorical variables were converted into numerical form using label encoding and one-hot encoding.

• Binary variables were label encoded, while multi-category variables were one-hot encoded.

• Loan_ID is just and identifier So, column (Loan_ID) was removed.

• The dataset now contains only numeric features and is ready for machine learning model training.
